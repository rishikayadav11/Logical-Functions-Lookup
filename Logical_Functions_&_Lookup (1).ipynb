4. Find the **Department** of Employee ID = E103. Also explain which LOOKUP function is used and why?

   | **EmpID** | **Name** | **Department** |
   |-----------|----------|----------------|
   |   E101    |   Raju   |   IT           |
   |   E102    |   Riya   |   HR           |
   |   E103    |   Roshni |   Finance      |
   |   E104    |   Sejal  |   Sales        |

Ans- I have used the **VLOOKUP** function.

- **FORMULA:** =VLOOKUP(Lookup value/cell, Table range, Column index no., FALSE)

    Assuming table is in the range **A1:C5**, the formula is:
               
     =VLOOKUP(A4, A1:C5, 3, False)

- **RESULT:** Finance

- **Why it is used:** It is the classic, widely known function. It searches for the EmpID in the very first column of your selected range (A1:C5), moves 3 columns to the right to find the Department, and uses FALSE to ensure it only accepts an exact match. Note that with VLOOKUP, your search column must always be the leftmost column in the range.

5.  A company has set the following **commission structure** for its sales team:

    | **Sales Amount** | **Commission Rate** |
    |------------------|---------------------|
    |   0              |   5%                |
    |   20000          |   10%               |
    |   50000          |   15%               |
    |   80000          |   20%               |

    If a salesperson achieves sales of ₹ 45,000, use the **VLOOKUP function** to determine the commission rate applicable.

Ans-

- **FORMULA:** =VLOOKUP(Lookup value, Table range, Column index no., TRUE)

  Assuming table is in the range **A1:B5**, the formula is:
   
  =VLOOKUP(45000, A1:B5, 2, TRUE)

- **RESULT:** 10%

- **How it works:**

   **Lookup Value:** 45,000.

   **Table Array/Range:** A1:B5 (The range containing sales tiers and rates).

   **Col_index_num:** 2 (the second column holds the rates).

   **Range_lookup:** TRUE (or 1) tells Excel to find the largest value in the first column that is less than or equal to the lookup value. Since 45,000 is greater than 20,000 but less than 50,000, it selects the rate for 20,000.
